```python
from pyEDA.main import *
from scipy.stats import skew
from scipy.stats import kurtosis
datadb_processed = {}
for k,v in data.items():
    m, wd, eda_clean = process_statistical(v[0], use_scipy=True, sample_rate=700, new_sample_rate=100, segment_width=1, segment_overlap=0)
    label_downsampled = []
    for i in range(0,len(v[1])-700,700):
        #label_downsampled.append(np.max(v[1][i:i+7000]))
        label_downsampled.append(stats.mode(v[1][i:i+700], axis=None, keepdims=False).mode)
    x_mean1 = np.mean(eda_clean[:-1],axis=-1)
    x_mean = np.mean(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_var = np.var(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_stddev = np.std(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_min = np.min(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_max = np.max(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_median = np.median(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_skew = skew(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x_kurtosis = kurtosis(wd['filtered_phasic_gsr'][:-1],axis=-1)
    x = np.vstack((x_mean1,x_mean,x_var,x_stddev,x_min,x_max,x_median,x_skew,x_kurtosis))
    print(x.shape)
    print(len(label_downsampled))
    datadb_processed[k] = [x,label_downsampled[:x_mean.shape[0]]]
```

```python
files_list = list(datadb_processed.keys())
file_list_num = np.arange(len(files_list))
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn import preprocessing
kf = KFold(n_splits=5)
#file_list_num
for i, (train_index, test_index) in enumerate(kf.split(file_list_num)):
    print(f"Fold {i}:")
    #print(f"  Train: index={train_index}")
    #print(f"  Test:  index={test_index}")
    train_db_x = []
    train_db_y = []
    test_db_x = []
    test_db_y = []
    for tr in train_index:
        #print(datadb_processed[files_list[tr]][0].shape)
        train_db_x.append(datadb_processed[files_list[tr]][0])
        train_db_y.append(datadb_processed[files_list[tr]][1])
    for tr in test_index:
      test_db_x.append(datadb_processed[files_list[tr]][0])
      test_db_y.append(datadb_processed[files_list[tr]][1])

    train_X = np.concatenate(train_db_x,axis=1)
    train_Y = np.concatenate(train_db_y)
    test_X = np.concatenate(test_db_x,axis=1)
    test_Y = np.concatenate(test_db_y)
    #print(train_X.T.shape)


    m = OneVsRestClassifier(SVC(kernel='rbf')).fit(train_X.T, train_Y.reshape(-1,1))
    predict_Y = m.predict(test_X.T)
    print(classification_report(test_Y.reshape(-1,1),predict_Y))
```
